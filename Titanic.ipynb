```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
```

```python
df=pd.read_csv("Desktop\\py-master\\py-master\\ML\\9_decision_tree\\Exercise\\titanic.csv")
df.head()
```

```python
X=df.drop(["Survived",'Name','PassengerId','Cabin','Ticket'],axis=1)
y=df.Survived
```

```python
X.head()
```

```python
X.Age.fillna(X.Age.median(),inplace=True)
X.Fare.fillna(X.Fare.median(),inplace=True)
X.Embarked.fillna(method='ffill',inplace=True)
```

```python
y.head()
```

```python
le=LabelEncoder()
```

```python
X["Sex_n"]=le.fit_transform(X['Sex'])
X["SibSp_n"]=le.fit_transform(X['SibSp'])
X["Parch_n"]=le.fit_transform(X['Parch'])
X["Embarked_n"]=le.fit_transform(X['Embarked'])
```

```python
X.head()
```

```python
X.drop(['Sex','Parch','SibSp','Embarked'],axis=1,inplace=True)
```

```python
X.head()
```

## Decision Tree Classifier

```python
model=DecisionTreeClassifier()
```

```python
X_train, X_test, y_train, y_test =train_test_split(X,y,test_size=0.25)
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

## Logistic Regression

```python
mod2=LogisticRegression()
```

```python
mod2.fit(X_train,y_train)
```

```python
mod2.score(X_test,y_test)
```

## Support Vector Machines

```python
mod3=SVC()
```

```python
mod3.fit(X_train,y_train)
```

```python
mod3.score(X_test,y_test)
```

## Random Forest Classifier

```python
mod4=RandomForestClassifier(n_estimators=100)
```

```python
mod4.fit(X_train,y_train)
```

```python
mod4.score(X_test,y_test)
```

## Naive Bayes

```python
mod5=GaussianNB()
```

```python
mod5.fit(X_train,y_train)
```

```python
mod5.score(X_test,y_test)
```

From the analysis we conclude that Random Forest Classifier fits the model best.
